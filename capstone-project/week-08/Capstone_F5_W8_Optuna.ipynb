## ⚠️ F5 Boundary Saturation — What It Is and Why It Matters

F5 has converged to a boundary region: every high-value submission since W2
has X2, X3, X4 at or near 1.0. The W7 ATB is `[0.938, 1.0, 1.0, 1.0]`.

**The saturation problem:** when `best_point` has X2=X3=X4=1.0 and we generate
exploit candidates via `np.clip(best_point + N(0, σ), 0, 1)`, any noise that
pushes above 1.0 is hard-clipped back to exactly 1.0. The result is an asymmetric
distribution — X2/X3/X4 can drift *down* from 1.0 but never *up*. The RF
classifier, trained on data where high-y = X2/X3/X4 ≈ 1.0, then kills anything
that drifted down, leaving a filtered set where X2=X3=X4=1.0 on nearly every row.
The GP has zero variance on those three dimensions and pushes X1 to its
highest-uncertainty boundary. **Result: submission [1.0, 1.0, 1.0, 1.0]** —
a degenerate answer that gives no new information about X1.

**Why [1,1,1,1] is problematic:** W6 was `[0.781, 1.0, 1.0, 1.0]` = 5875.
W7 was `[0.938, 1.0, 1.0, 1.0]` = 7596. The trajectory is clearly X1→1.0.
But simply submitting [1,1,1,1] wastes the submission — it either confirms
the trend (no new information) or hits the true boundary and leaves the
question of whether X1=0.95 vs X1=1.0 is better unanswered.

**The fix:** see Step 8 for the boundary-aware candidate split.

## ⚙️ STEP 0 — F5 Configuration & Hyperparameter Documentation

### 🔧 F5-Specific Settings (Week 8 — OPTUNA VARIANT + BOUNDARY FIX)

**Three changes** vs the standard W8 pipeline:

**1. Sobol sequences** replace `np.random.uniform` for the 1,000 exploration candidates.
Implementation: `scipy.stats.qmc.Sobol(d=4, scramble=True)` — draw next power of 2,
trim to 1,000. In 4D, Sobol's space-filling guarantee is more meaningful than
random draws, which cluster by chance.

**2. Dual kernel GP** — Matérn 5/2 and RBF fitted sequentially on identical training data.
UCB scores computed from each. Best candidate across both kernels submitted.
EI is primary selector (function has real signal at 7596); UCB pooling used
as the dual-kernel mechanism. If EI is positive, EI selects; UCB is reference.

**3. Boundary-aware candidate split (F5-specific)** — addresses a structural issue
where `best_point = [0.938, 1.0, 1.0, 1.0]` causes boundary saturation.
Without intervention, all exploit candidates have X2=X3=X4=1.0 (clipped)
and the GP returns the degenerate submission [1.0, 1.0, 1.0, 1.0].

The exploit pool (9,000 candidates) is split:
- **50% (4,500)** — standard Gaussian cloud, X2/X3/X4 clipped to 1.0 as expected
- **50% (4,500)** — X1-targeted probe: X1 ~ Uniform[0.90, 1.0], X2=X3=X4 hard-locked to 1.0

This forces the GP to see real X1 variance across the boundary-locked region,
producing a meaningful X1 answer rather than the degenerate all-ones result.
The `F5_X1_PROBE = True` flag controls this behaviour.

All other settings (TOP_PERCENTILE=30, FILTER_PERCENTILE=50, GP_RESTARTS=6,
EXPLOIT_RATIO=0.9, EXPLOIT_SIGMA=0.035, UCB_KAPPA=2.0) are unchanged.

## 🏷️ STEP 4 — Binary Classification Labels

**Why classify?** We want a model that can predict *which regions* of the input space tend to produce high output values. This is a classification task: class 1 = "good" (top 30%), class 0 = "bad" (bottom 70%).

**MAXIMIZATION direction**: class 1 = **highest** values. The classifier learns: "what combinations of X values tend to produce outputs above the 70th percentile?"

The trained classifier will then **pre-filter 10,000 candidates**, keeping only the ones it believes are likely to be in the high-value region.

**What happens without binary classification?**
Without a classifier pre-filter, the GP would receive all 10,000 random/Gaussian candidates — the vast majority of which are low-value points. In an *n*-sample, *d*-dimensional space this matters because:

- The GP's uncertainty is highest far from observed data — low-value regions get high σ and *accidentally* high EI
- Without filtering, the acquisition function wastes evaluations probing regions the classifier already knows are poor
- A good classifier concentrates the GP's attention on the ~30% of space that matters, acting as a **learned search space reduction** step before the expensive GP computation

In short: the classifier does not replace the GP — it *narrows the candidate pool* so the GP only has to decide *where exactly* within the high-value region to query next.

**Week-on-week CNN tracking:** F5's extreme value range (0.1 to 2912) gives the CNN the strongest class signal in the W6 batch. At n=24 in 4D the Conv1d kernel scanning pairs of boundary coordinates should detect the high-value pattern clearly. Compare with F4 (n=34, 4D) to isolate the pure sample-size effect on CNN accuracy.

````python
import torch
import torch.nn as nn
import torch.optim as optim_torch
from sklearn.base import BaseEstimator, ClassifierMixin

# ── CNN Wrapper — sklearn-compatible for use in cross_val_score ───────────
class TinyCNNClassifier(BaseEstimator, ClassifierMixin):
    """
    Sklearn-compatible wrapper around a tiny 1D CNN.
    Added as Model 8 — a learning exercise to track CNN performance
    week on week as sample size grows from 15 toward 40+.
    Architecture: Conv1d(1->8, k=2) -> ReLU -> Dropout(0.3) -> Linear(8->1)
    Why tiny: with n=15, any larger CNN will overfit immediately.
    Why included: to document empirically where CNN beats MLP (if ever),
    and to demonstrate Module 17 CNN concepts in the BBO context.
    """
    def __init__(self, n_epochs=200, lr=0.01, seed=42):
        self.n_epochs = n_epochs
        self.lr       = lr
        self.seed     = seed

    def fit(self, X, y):
        torch.manual_seed(self.seed)
        n_feat = X.shape[1]
        self.classes_ = np.unique(y)

        class _CNN(nn.Module):
            def __init__(self):
                super().__init__()
                self.conv = nn.Conv1d(1, 8, kernel_size=min(2, n_feat), padding=0)
                self.relu = nn.ReLU()
                self.drop = nn.Dropout(0.3)
                conv_out  = 8 * max(1, n_feat - min(2, n_feat) + 1)
                self.fc   = nn.Linear(conv_out, 1)
            def forward(self, x):
                x = self.relu(self.conv(x))
                x = self.drop(x.view(x.size(0), -1))
                return self.fc(x).squeeze(-1)

        self.model_ = _CNN()
        n_pos = y.sum(); n_neg = len(y) - n_pos
        pw    = torch.tensor([n_neg / max(n_pos, 1)], dtype=torch.float)
        loss_fn = nn.BCEWithLogitsLoss(pos_weight=pw)
        opt   = optim_torch.Adam(self.model_.parameters(), lr=self.lr,
                                 weight_decay=1e-3)
        Xt = torch.FloatTensor(X).unsqueeze(1)
        yt = torch.FloatTensor(y.astype(float))
        self.model_.train()
        for _ in range(self.n_epochs):
            opt.zero_grad()
            loss_fn(self.model_(Xt), yt).backward()
            opt.step()
        return self

    def predict_proba(self, X):
        self.model_.eval()
        with torch.no_grad():
            logits = self.model_(torch.FloatTensor(X).unsqueeze(1))
            p = torch.sigmoid(logits).numpy()
        return np.column_stack([1 - p, p])

    def predict(self, X):
        return (self.predict_proba(X)[:, 1] >= 0.5).astype(int)

# ── Model registry ────────────────────────────────────────────────────────
cv = StratifiedKFold(n_splits=n_cv_folds, shuffle=True, random_state=RANDOM_SEED)

models = {
    'Linear SVM'         : SVC(kernel='linear', C=1.0, probability=True, random_state=RANDOM_SEED),
    'Decision Tree'      : DecisionTreeClassifier(max_depth=4, random_state=RANDOM_SEED),
    'Random Forest'      : RandomForestClassifier(n_estimators=100, random_state=RANDOM_SEED),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=RANDOM_SEED),
    f'NN-Small {NN_SMALL}' : MLPClassifier(hidden_layer_sizes=NN_SMALL, max_iter=500,
                                            early_stopping=True, validation_fraction=0.15,
                                            random_state=RANDOM_SEED),
    f'NN-Medium {NN_MEDIUM}': MLPClassifier(hidden_layer_sizes=NN_MEDIUM, max_iter=500,
                                            early_stopping=True, validation_fraction=0.15,
                                            random_state=RANDOM_SEED),
    f'NN-Large {NN_LARGE}' : MLPClassifier(hidden_layer_sizes=NN_LARGE, max_iter=500,
                                            early_stopping=True, validation_fraction=0.15,
                                            random_state=RANDOM_SEED),
    'CNN-1D (learning)'  : TinyCNNClassifier(n_epochs=200, lr=0.01, seed=RANDOM_SEED),
}

cv_results = {}
print('='*70)
print(f'  {n_cv_folds}-FOLD CV -- 8 MODELS -- {FUNCTION_ID}  ({n_samples} samples, {n_dims}D)')
print(f'  Model 8: CNN-1D added as learning exercise -- tracking week on week')
print('='*70)
print(f'  {"Model":<28} {"Mean CV":>9} {"+-Std":>8}   Folds')
print('  ' + '-'*65)

for name, model in models.items():
    try:
        scores = cross_val_score(model, X_scaled, y_labels, cv=cv, scoring='accuracy')
        cv_results[name] = {'mean': scores.mean(), 'std': scores.std(),
                             'scores': scores, 'model': model}
        fold_str = '  '.join([f'{s:.3f}' for s in scores])
        flag = '★' if scores.mean() == max(r['mean'] for r in cv_results.values()) else ' '
        print(f'  {flag} {name:<27} {scores.mean():>8.1%} {scores.std():>7.1%}   [{fold_str}]')
    except Exception as e:
        cv_results[name] = {'mean': 0.0, 'std': 1.0,
                             'scores': np.zeros(n_cv_folds), 'model': model}
        print(f'  ✗ {name:<27} FAILED: {str(e)[:40]}')

ranked_models = sorted(cv_results.items(), key=lambda x: x[1]['mean'], reverse=True)
best_name = ranked_models[0][0]
best_info = ranked_models[0][1]

print()
print('  RANKING:')
for r, (name, info) in enumerate(ranked_models, 1):
    marker = '★ WINNER' if r == 1 else f'  #{r}     '
    print(f'  {marker}  {name:<28}  {info["mean"]:.1%} ± {info["std"]:.1%}')

print(f'\n  ✅ SELECTED: {best_name}  (CV={best_info["mean"]:.1%})')
````

## 🎲 STEP 8 — Generate & Filter Candidates (Optuna: Sobol + F5 Boundary-Aware Split)

### Module 18 + Optuna Context
Module 18 identifies random search as the third tier of hyperparameter optimisation
strategies. Optuna's NeurIPS 2020 contribution shows that **quasi-random Sobol sequences**
outperform pure random draws for candidate initialisation — particularly as dimension
increases. Each Sobol point is placed into the largest remaining gap in the space,
guaranteeing coverage. Random draws cluster by chance.

**F5 is 4D.** Sobol is applied to the 1,000 explore candidates, giving even [0,1]⁴
coverage as a safety net against boundary overfitting.

---

### ⚠️ F5 Boundary Saturation — Technical Detail

**Standard pipeline behaviour (problematic for F5):**
```
X_exploit = np.clip(best_point + N(0, 0.035), 0, 1)
```
With `best_point = [0.938, 1.0, 1.0, 1.0]`:
- X2/X3/X4: noise can go *down* from 1.0 but is clipped at 1.0 going up
- Distribution is left-tailed: most values sit at exactly 1.0, tail extends to ~0.85
- RF classifier (trained on high-y ↔ X2/X3/X4≈1.0) kills anything that drifted down
- After filtering: nearly all candidates have X2=X3=X4=1.0
- GP sees zero variance on X2/X3/X4 → treats them as constants
- GP + UCB maximises X1 uncertainty → X1 → 1.0
- **Submission: [1.0, 1.0, 1.0, 1.0]** — degenerate, no new X1 information

**Fixed pipeline (this notebook):**

| Pool | Size | X1 | X2 | X3 | X4 |
|------|------|----|----|----|----|  
| Gaussian exploit | 4,500 | N(0.938, 0.035) | clipped→1.0 | clipped→1.0 | clipped→1.0 |
| X1 probe exploit | 4,500 | U[0.90, 1.0] | 1.0 (locked) | 1.0 (locked) | 1.0 (locked) |
| Sobol explore    | 1,000 | Sobol[0,1] | Sobol[0,1] | Sobol[0,1] | Sobol[0,1] |

The X1-probe candidates explicitly cover the [0.90, 1.0] range with X2=X3=X4=1.0
locked. This gives the GP real X1 signal in the confirmed-best region, and the
submission will be a meaningful X1 value rather than the all-ones boundary.

**Why this is the right decision:** the F5 trajectory is clearly X1→1.0 over time
(W6: X1=0.781 → W7: X1=0.938). The question is whether the true optimum is at
X1=1.0 or somewhere in [0.95, 1.0]. The boundary-aware split answers this directly.

````python
best_clf = trained_models[best_name]
n_exploit = int(N_CANDIDATES * EXPLOIT_RATIO)
n_explore = N_CANDIDATES - n_exploit

# ── F5 BOUNDARY SATURATION NOTE ──────────────────────────────────
# best_point = [0.938, 1.0, 1.0, 1.0] — X2/X3/X4 already at boundary.
# np.clip collapses all exploit noise above 1.0 back to 1.0.
# This is CORRECT — X2/X3/X4=1.0 is confirmed best region.
# Real unknown: optimal X1. Strategy: half of exploit probes X1 near 1.0,
# half probes X1 freely. Sobol explore covers all of [0,1]^4.

# ── Exploit split: 50% precision X1 probe, 50% standard Gaussian ─
n_exploit_probe  = n_exploit // 2
n_exploit_gauss  = n_exploit - n_exploit_probe

# Standard Gaussian cloud (X2/X3/X4 clip to 1.0 — expected)
X_exploit_gauss  = np.clip(best_point + np.random.normal(0, EXPLOIT_SIGMA, (n_exploit_gauss, n_dims)), 0, 1)

# X1-targeted probe: X1 ~ Uniform[0.90, 1.0], X2/X3/X4 = 1.0 (boundary lock)
X_exploit_probe  = np.ones((n_exploit_probe, n_dims))
X_exploit_probe[:, 0] = np.random.uniform(0.90, 1.0, n_exploit_probe)
print(f'  F5 boundary probe: {n_exploit_probe:,} candidates with X2=X3=X4=1.0, X1~U[0.90,1.0]')

X_exploit = np.vstack([X_exploit_gauss, X_exploit_probe])

# ── Explore: Sobol sequences (full [0,1]^4 coverage) ─────────────
if USE_SOBOL:
    import math
    n_sobol_draw = 2 ** math.ceil(math.log2(max(n_explore, 2)))
    sobol_eng    = Sobol(d=n_dims, scramble=True, seed=RANDOM_SEED)
    X_explore    = sobol_eng.random(n_sobol_draw)[:n_explore]
    print(f'  Sobol: drew {n_sobol_draw} → trimmed to {n_explore}')
else:
    X_explore = np.random.uniform(0, 1, (n_explore, n_dims))

X_cands  = np.vstack([X_exploit, X_explore])
src_tag  = np.array(['exploit']*n_exploit + ['explore']*n_explore)

X_cands_sc   = scaler.transform(X_cands)
proba_all    = best_clf.predict_proba(X_cands_sc)[:, 1]
keep_mask    = proba_all >= np.percentile(proba_all, FILTER_PERCENTILE)
X_filt       = X_cands[keep_mask]
proba_filt   = proba_all[keep_mask]
src_filt     = src_tag[keep_mask]

print('═'*65)
print('  CANDIDATE GENERATION — F5 BOUNDARY-AWARE (OPTUNA VARIANT)')
print('═'*65)
print(f'  Generated  : {N_CANDIDATES:,}')
print(f'    Gaussian exploit : {n_exploit_gauss:,}  (X2/X3/X4 clip to 1.0 — correct)')
print(f'    X1 probe exploit : {n_exploit_probe:,}  (X1~U[0.90,1.0], X2=X3=X4=1.0)')
print(f'    Sobol explore    : {n_explore:,}  (full [0,1]⁴ coverage)')
print(f'  Filtered by: {best_name}')
print(f'  Kept       : {len(X_filt):,} (top {100-FILTER_PERCENTILE}% by P(class=1))')
print(f'  Exploit %  : {100*np.mean(src_filt=="exploit"):.1f}%')
print(f'  X1 range in X_filt: [{X_filt[:,0].min():.3f}, {X_filt[:,0].max():.3f}]')
print(f'  → {len(X_filt):,} candidates passed to Dual Gaussian Process')
````

## 🔬 STEP 9 — Dual Kernel Gaussian Process (Optuna: Multi-Kernel Hedging)

### Optuna + Module 18 Context
Standard GP-BO uses a single kernel encoding a fixed smoothness assumption.
The Optuna NeurIPS 2020 paper uses **multi-kernel Thompson sampling**: multiple
kernels are fitted simultaneously and the best candidate across all kernels is
selected. This hedges against unknown smoothness.

**Two kernels for F5:**
- **Matérn 5/2** — allows sharp transitions. Appropriate if the F5 peak near
  X1=1.0 is a narrow spike at the boundary.
- **RBF** — assumes smooth landscape. Appropriate if F5 increases monotonically
  toward the boundary with no sharp cutoff at X1=1.0.

**Interaction with the boundary fix:** the X1-probe candidates from Step 8 give
both GP kernels real X1 variance to fit against in the [0.90, 1.0] region.
Without the boundary fix, both kernels would see a near-degenerate candidate set
(X1≈0.95, X2=X3=X4=1.0 everywhere) and converge to the same [1,1,1,1] answer.
With the fix, Matérn and RBF may disagree on the optimal X1 value — exactly
the scenario where dual-kernel pooling adds value.

**Implementation:** both GPs fitted sequentially on the same scaled training data.
`gp = gp_matern` for all downstream visualisation steps. Dual-kernel selection
happens at Step 10.

````python
# ── Kernel 1: Matern 5/2 (standard) ─────────────────────────
kernel_matern = C(1.0, (1e-3, 1e3)) * Matern(length_scale=1.0, length_scale_bounds=(1e-2, 1e2), nu=2.5)
gp_matern = GaussianProcessRegressor(kernel=kernel_matern, alpha=1e-6, normalize_y=True,
                                      n_restarts_optimizer=GP_RESTARTS, random_state=RANDOM_SEED)
gp_matern.fit(X_scaled, y_train)

# ── Kernel 2: RBF (Optuna addition) ───────────────────────────
kernel_rbf = C(1.0, (1e-3, 1e3)) * RBF(length_scale=1.0, length_scale_bounds=(1e-2, 1e2))
gp_rbf = GaussianProcessRegressor(kernel=kernel_rbf, alpha=1e-6, normalize_y=True,
                                   n_restarts_optimizer=GP_RESTARTS, random_state=RANDOM_SEED)
gp_rbf.fit(X_scaled, y_train)

# gp = gp_matern for all downstream single-GP steps (viz, sweeps, surfaces)
gp = gp_matern

def gp_r2_rmse(gp_model, X_sc, y):
    y_pred, _ = gp_model.predict(X_sc, return_std=True)
    ss_res = np.sum((y - y_pred)**2)
    ss_tot = np.sum((y - y.mean())**2)
    r2   = 1 - ss_res/ss_tot if ss_tot > 0 else 0.0
    rmse = np.sqrt(np.mean((y - y_pred)**2))
    return r2, rmse

r2_m, rmse_m = gp_r2_rmse(gp_matern, X_scaled, y_train)
r2_r, rmse_r = gp_r2_rmse(gp_rbf,    X_scaled, y_train)
r2, rmse     = r2_m, rmse_m

X_filt_sc = scaler.transform(X_filt)
mu_matern, sigma_matern = gp_matern.predict(X_filt_sc, return_std=True)
mu_rbf,    sigma_rbf    = gp_rbf.predict(X_filt_sc,    return_std=True)
mu, sigma = mu_matern, sigma_matern

print('═'*65)
print('  DUAL KERNEL GP (OPTUNA VARIANT)')
print('═'*65)
print(f'  Matern 5/2 : {gp_matern.kernel_}')
print(f'    R²={r2_m:.4f}  RMSE={rmse_m:.2e}  μ=[{mu_matern.min():.4f},{mu_matern.max():.4f}]  σ=[{sigma_matern.min():.4f},{sigma_matern.max():.4f}]')
print(f'  RBF        : {gp_rbf.kernel_}')
print(f'    R²={r2_r:.4f}  RMSE={rmse_r:.2e}  μ=[{mu_rbf.min():.4f},{mu_rbf.max():.4f}]  σ=[{sigma_rbf.min():.4f},{sigma_rbf.max():.4f}]')
print(f'  Downstream viz uses: Matern 5/2 (gp = gp_matern)')
````

## 🎯 STEP 10 — Acquisition Functions: EI & UCB (Optuna: Dual Kernel Pooling)

EI and UCB are computed from **both kernels** independently.

**Selection logic — EI primary:**
F5 has real signal (ATB 7596.79) — EI is expected to be positive and is the
primary selector. The best EI candidate across both kernels (Matérn and RBF)
is submitted. If EI is zero across all candidates (unexpected for F5), UCB
is used as a fallback.

**Boundary fix interaction:** the X1-probe candidates (X1~U[0.90,1.0],
X2=X3=X4=1.0) that survived the RF filter will be visible to EI here.
EI = (μ - y_best - ξ)·Φ(Z) + σ·φ(Z). With y_best=7596.79, EI will be
positive only for candidates the GP predicts might beat that value.
The X1-probe candidates give EI a fair chance to find the optimal X1
in [0.90, 1.0] rather than defaulting to X1=1.0 through pure UCB pressure.

**EI formula:**
$$EI(x) = (\mu(x) - y_{best} - \xi)\cdot\Phi(Z) + \sigma(x)\cdot\phi(Z)$$

**UCB formula:** $UCB(x) = \mu(x) + 2.0 \cdot \sigma(x)$

Both computed for MAXIMIZATION. Dual-kernel pool: best EI across Matérn and
RBF independently evaluated on the same filtered candidate set from Step 8.

````python
# ── EI from Matern ───────────────────────────────────────────
imp_m     = mu_matern - best_value - EI_XI
Z_m       = imp_m / (sigma_matern + 1e-9)
EI_matern = np.maximum(imp_m * norm.cdf(Z_m) + sigma_matern * norm.pdf(Z_m), 0)
exploit_term_m = imp_m * norm.cdf(Z_m)
explore_term_m = sigma_matern * norm.pdf(Z_m)

# ── EI from RBF ───────────────────────────────────────────────
imp_r  = mu_rbf - best_value - EI_XI
Z_r    = imp_r / (sigma_rbf + 1e-9)
EI_rbf = np.maximum(imp_r * norm.cdf(Z_r) + sigma_rbf * norm.pdf(Z_r), 0)

# ── UCB from both kernels ─────────────────────────────────────
UCB_matern = mu_matern + UCB_KAPPA * sigma_matern
UCB_rbf    = mu_rbf    + UCB_KAPPA * sigma_rbf

# ── Pool: best candidate across both kernels ──────────────────
best_ucb_m_idx = np.argmax(UCB_matern)
best_ucb_r_idx = np.argmax(UCB_rbf)
best_ei_m_idx  = np.argmax(EI_matern)
best_ei_r_idx  = np.argmax(EI_rbf)

# Combined EI pool: best EI across both kernels
if EI_rbf[best_ei_r_idx] >= EI_matern[best_ei_m_idx]:
    best_ei_pool_idx    = best_ei_r_idx
    best_ei_pool_kernel = 'RBF'
else:
    best_ei_pool_idx    = best_ei_m_idx
    best_ei_pool_kernel = 'Matern'

# Combined UCB pool
if UCB_rbf[best_ucb_r_idx] >= UCB_matern[best_ucb_m_idx]:
    best_ucb_pool_idx    = best_ucb_r_idx
    best_ucb_pool_kernel = 'RBF'
    best_ucb_pool_val    = UCB_rbf[best_ucb_r_idx]
else:
    best_ucb_pool_idx    = best_ucb_m_idx
    best_ucb_pool_kernel = 'Matern'
    best_ucb_pool_val    = UCB_matern[best_ucb_m_idx]

# ── Select submission (EI primary) ──────────────────────────────────
ei_max_pool = max(EI_matern.max(), EI_rbf.max())
# EI primary — switch to UCB if EI completely flat
if ei_max_pool > 1e-12:
    winner_idx    = best_ei_pool_idx
    winner_kernel = best_ei_pool_kernel
    winner_source = 'EI'
else:
    winner_idx    = best_ucb_pool_idx
    winner_kernel = best_ucb_pool_kernel
    winner_source = 'UCB (EI=0 fallback)'

submission = X_filt[winner_idx]

# Expose winner mu/sigma from winning kernel
if winner_kernel == 'RBF':
    sub_mu    = mu_rbf[winner_idx]
    sub_sigma = sigma_rbf[winner_idx]
    sub_ucb   = UCB_rbf[winner_idx]
    sub_ei    = float(EI_rbf[winner_idx])
else:
    sub_mu    = mu_matern[winner_idx]
    sub_sigma = sigma_matern[winner_idx]
    sub_ucb   = UCB_matern[winner_idx]
    sub_ei    = float(EI_matern[winner_idx])

# Expose standard names for downstream cells (Matern as reference)
EI_acq       = EI_matern
UCB_acq      = UCB_matern
sub_exp      = float(exploit_term_m[winner_idx])
sub_expl     = float(explore_term_m[winner_idx])

print('═'*65)
print('  ACQUISITION — DUAL KERNEL POOL (MAXIMIZE ⬆️)')
print('═'*65)
print(f'  EI  Matern : max={EI_matern.max():.4e}   EI>0: {100*np.mean(EI_matern>0):.1f}%')
print(f'  EI  RBF    : max={EI_rbf.max():.4e}   EI>0: {100*np.mean(EI_rbf>0):.1f}%')
print(f'  UCB Matern : max={UCB_matern.max():.4e}')
print(f'  UCB RBF    : max={UCB_rbf.max():.4e}')
print()
print(f'  Pool EI winner  : {best_ei_pool_kernel}  idx={best_ei_pool_idx}')
print(f'  Pool UCB winner : {best_ucb_pool_kernel}  UCB={best_ucb_pool_val:.4e}')
print()
print(f'  ★ SELECTED via {winner_source} — kernel: {winner_kernel}')
print(f'  ★ SUBMISSION : {submission}')
print(f'  μ={sub_mu:.6e}  σ={sub_sigma:.6e}  UCB={sub_ucb:.6e}  EI={sub_ei:.6e}')
print(f'  Delta vs best: {sub_mu - best_value:+.6e}')
status = '✅ PREDICTS NEW BEST' if sub_mu > best_value else '✅ PREDICTS RECOVERY' if sub_mu > latest_val else '🔍 EXPLORING'
print(f'  Status: {status}')
print(f'  Optuna dual kernel: {winner_kernel} won — hedges unknown smoothness assumption')
````

## 🗺️ STEP 12A — GP Acquisition Surface

For the two most sensitive dimensions (found by GP gradient analysis), we plot 2D contour surfaces of μ, EI and UCB. All other dimensions are fixed at the best-known values.

This gives a direct spatial view of where the GP thinks the landscape peaks and where the acquisition functions want us to go next.

````python
# ══════════════════════════════════════════════════════════════════
# STEP 12A — GP Acquisition Surfaces  (standard for all functions)
# ══════════════════════════════════════════════════════════════════
delta = 0.01
sensitivities = []
for d in range(n_dims):
    p_up = best_point.copy(); p_up[d] = min(1.0, best_point[d]+delta)
    p_dn = best_point.copy(); p_dn[d] = max(0.0, best_point[d]-delta)
    m_up, _ = gp.predict(scaler.transform(p_up.reshape(1,-1)), return_std=True)
    m_dn, _ = gp.predict(scaler.transform(p_dn.reshape(1,-1)), return_std=True)
    sensitivities.append(abs(float(m_up) - float(m_dn)) / (2*delta))
top2 = np.argsort(sensitivities)[::-1][:2]
d1, d2 = top2[0], top2[1]
print(f'  Top 2 sensitive dims: X{d1+1} ({sensitivities[d1]:.4f}), X{d2+1} ({sensitivities[d2]:.4f})')

# Build grid — extend slightly past 1.0 so boundary markers are visible
g_res = 45
g1 = np.linspace(0, 1, g_res)
g2 = np.linspace(0, 1, g_res)
G1, G2 = np.meshgrid(g1, g2)
grid_pts = np.tile(best_point, (g_res*g_res, 1))
grid_pts[:, d1] = G1.ravel(); grid_pts[:, d2] = G2.ravel()
grid_sc = scaler.transform(grid_pts)
mg, sg = gp.predict(grid_sc, return_std=True)
ig = mg - best_value - EI_XI; Zg = ig/(sg+1e-9)
EIg = np.maximum(ig*norm.cdf(Zg) + sg*norm.pdf(Zg), 0)
UCBg = mg + UCB_KAPPA*sg

fig, axes = plt.subplots(1, 3, figsize=(18, 5.5))
fig.suptitle(
    f'{FUNCTION_ID} Week {WEEK} — GP Acquisition Surfaces (X{d1+1} x X{d2+1})\n'
    f'Other dims fixed at best point  |  '
    f'★ best=({best_point[d1]:.3f}, {best_point[d2]:.3f})  '
    f'◆ submit=({submission[d1]:.3f}, {submission[d2]:.3f})',
    fontsize=11, fontweight='bold')

_bx, _by = float(best_point[d1]), float(best_point[d2])
_sx, _sy = float(submission[d1]), float(submission[d2])

# Nudge markers that sit exactly on the boundary inward so they're visible
_PAD = 0.018
def _nudge(v): return max(_PAD, min(1.0 - _PAD, v))
_bx_plot, _by_plot = _nudge(_bx), _nudge(_by)
_sx_plot, _sy_plot = _nudge(_sx), _nudge(_sy)

for ax, Z_grid, title, cmap, clabel in [
    (axes[0], mg.reshape(g_res, g_res),   'GP Mean μ',           'viridis', 'μ'),
    (axes[1], EIg.reshape(g_res, g_res),  'Expected Improvement', 'plasma',  'EI'),
    (axes[2], UCBg.reshape(g_res, g_res), f'UCB (μ + {UCB_KAPPA}·σ)', 'magma', 'UCB'),
]:
    cf = ax.contourf(G1, G2, Z_grid, levels=25, cmap=cmap, alpha=0.88)
    ax.contour(G1, G2, Z_grid, levels=8, colors='white', alpha=0.2, linewidths=0.5)
    plt.colorbar(cf, ax=ax, label=clabel, pad=0.01)

    # Training points
    ax.scatter(X_train[:, d1], X_train[:, d2], c='white', s=45,
               edgecolors='black', linewidths=0.8, zorder=4, alpha=0.9, label='Train')

    # Full-span crosshairs at true best location
    ax.axvline(_bx, color='#ff4444', lw=1.5, ls='--', alpha=0.85, zorder=5)
    ax.axhline(_by, color='#ff4444', lw=1.5, ls='--', alpha=0.85, zorder=5)

    # Best point star — nudged inward if on boundary, outlined with contrasting colour
    ax.scatter(_bx_plot, _by_plot,
               c='red', s=500, marker='*', zorder=8,
               edgecolors='white', linewidths=2.5,
               label=f'★ Best ({_bx:.3f},{_by:.3f})')

    # Submission diamond — nudged inward if on boundary
    ax.axvline(_sx, color='#00ff88', lw=1.2, ls=':', alpha=0.7, zorder=5)
    ax.axhline(_sy, color='#00ff88', lw=1.2, ls=':', alpha=0.7, zorder=5)
    ax.scatter(_sx_plot, _sy_plot,
               c='#00ff88', s=280, marker='D', zorder=9,
               edgecolors='black', linewidths=1.5,
               label=f'◆ Submit ({_sx:.3f},{_sy:.3f})')

    ax.set_xlabel(f'X{d1+1}', fontsize=10)
    ax.set_ylabel(f'X{d2+1}', fontsize=10)
    ax.set_title(title, fontweight='bold', fontsize=11)
    ax.set_xlim(-0.01, 1.01)
    ax.set_ylim(-0.01, 1.01)
    ax.legend(fontsize=7.5, loc='best',
              framealpha=0.85, edgecolor='grey')

plt.tight_layout()
plt.savefig(f'{FUNCTION_ID}_W{WEEK}_Step12A_GPSurfaces.png', dpi=150, bbox_inches='tight')
plt.show()
print(f'Saved: {FUNCTION_ID}_W{WEEK}_Step12A_GPSurfaces.png')
````
